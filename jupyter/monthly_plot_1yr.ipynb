```python
import os
import glob
import numpy as np
import clima_anom as ca
import matplotlib.pyplot as plt
import matplotlib.cbook
import matplotlib.gridspec as gridspec
import cartopy
import cartopy.crs as ccrs
```

```python
yy = '2013'
```

```python
pre_dir = "/media/arturo/Arturo/Data/TRMM/3B42_v7_1dy/raw/*.nc4"
pre_list = glob.glob(pre_dir)
pre_list = sorted(pre_list)
len_list = len(pre_list)
print(pre_dir)
```

```python
tmp = ca.read_netcdf(list_01[0],1)
lat = tmp["lat"].data
lon = tmp["lon"].data

nlat = len(lat)
nlon = len(lon)

lon2d, lat2d = np.meshgrid(lon, lat)

print('Lat:',lat.shape)
print('Lon:',lon.shape)
```

```python
pr_mean = np.zeros([12,nlat,nlon])
pr_mean[0,:,:] = np.mean(pr01,axis=0)
pr_mean[1,:,:] = np.mean(pr02,axis=0)
pr_mean[2,:,:] = np.mean(pr03,axis=0)
pr_mean[3,:,:] = np.mean(pr04,axis=0)

pr_mean[4,:,:] = np.mean(pr05,axis=0)
pr_mean[5,:,:] = np.mean(pr06,axis=0)
pr_mean[6,:,:] = np.mean(pr07,axis=0)
pr_mean[7,:,:] = np.mean(pr08,axis=0)
pr_mean[8,:,:] = np.mean(pr09,axis=0)
pr_mean[9,:,:] = np.mean(pr10,axis=0)
pr_mean[10,:,:] = np.mean(pr11,axis=0)
pr_mean[11,:,:] = np.mean(pr12,axis=0)
```

```python
list_month = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
linewidth_shape = 0.2
levels = np.arange(0,60,4)

proj = ccrs.PlateCarree()
cmap = plt.cm.Spectral_r
# cmap = plt.cm.YlGnBu

fig = plt.figure(figsize=(10,8),dpi=200)
gs = gridspec.GridSpec(3,4)

a = 0
for row in range(3):
    for col in range(4):
        ax1 = plt.subplot(gs[row, col], projection = proj)
        ax1.coastlines(linestyle='-',linewidth=0.1,color='black',zorder=4)
        a1 = ax1.contourf(lon2d, lat2d,pr_mean[a,:,:],levels = levels, cmap=cmap, extend='max')
        ax1.text(0.1, 0.25, list_month[a], va='bottom', ha='center',rotation_mode='anchor',
        transform=ax1.transAxes)

        if a == 0:
            ax1.set_title('TRMM 3B42 v7',fontsize=8,loc='left')
            ax1.set_title('Daily',fontsize=8,loc='right')

        elif a == 3:
            ax1.set_title('Year: '+yy,fontsize=8,loc='right')

        if col == 0:
            gl = ax1.gridlines(crs=proj,draw_labels=True,linewidth=0.2,color='k',alpha=0.1,linestyle='--')
            gl.top_labels = False
            gl.bottom_labels = False
            gl.left_labels = True
            gl.right_labels = False
            gl.xlabel_style = {'size': 6, 'color': 'k'}
            gl.ylabel_style = {'size': 6, 'color': 'k'}

        if row == 2:
            gl = ax1.gridlines(crs=proj,draw_labels=True,linewidth=0.2,color='k',alpha=0.1,linestyle='--')
            gl.top_labels = False
            gl.bottom_labels = True
            gl.left_labels = False
            gl.right_labels = False
            gl.xlabel_style = {'size': 6, 'color': 'k'}
            gl.ylabel_style = {'size': 6, 'color': 'k'}

        else:
            gl = ax1.gridlines(crs=proj,draw_labels=True,linewidth=0.2,color='k',alpha=0.1,linestyle='--')
            gl.top_labels = False
            gl.bottom_labels = False
            gl.left_labels = False
            gl.right_labels = False
            gl.xlabel_style = {'size': 6, 'color': 'k'}
            gl.ylabel_style = {'size': 6, 'color': 'k'}    

        a = a + 1

plt.subplots_adjust(left=None, bottom=None, right=None, top=None, hspace=0.05, wspace=0.05)

cbaxes = fig.add_axes([0.127, 0.06, 0.772, 0.04]) 
fig.colorbar(a1, cax = cbaxes,orientation="horizontal")

salida = '../images/3B42_'+year+'_Monthly_Seasonal.png'
print(salida)
plt.savefig(salida,facecolor='white',bbox_inches ='tight',pad_inches = 0.1)
```
